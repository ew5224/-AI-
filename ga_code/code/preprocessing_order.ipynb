```python
import pandas as pd
import numpy
```

```python
order_ini = pd.read_csv("./data/order.csv")
cut_yield = pd.read_csv("./data/cut_yield.csv")
```

```python
order_ini
```

```python
cut_yield
```

```python
update_order = []
for row in order_ini.iterrows():
    
    month_index = 0
    if(row[1][0][5:7] == "04"):
        month_index = 0
    elif(row[1][0][5:7] == "05"):
        month_index = 1
    elif(row[1][0][5:7] == "06"):
        month_index = 2
    
    row[1][1] *= (100/float(cut_yield.loc[month_index, "BLK_1"]))*(1/0.975)
    row[1][2] *= (100/float(cut_yield.loc[month_index, "BLK_2"]))*(1/0.975)
    row[1][3] *= (100/float(cut_yield.loc[month_index, "BLK_3"]))*(1/0.975)
    row[1][4] *= (100/float(cut_yield.loc[month_index, "BLK_4"]))*(1/0.975)
    
    new_row = [row[1][0], row[1][1], row[1][2], row[1][3], row[1][4]]
    update_order.append(new_row)
    print(new_row)
#     print(cut_rate)
#     print(row[1][0][5:7], row[1][1], row[1][2], row[1][3], row[1][4])
```

```python
df = pd.DataFrame(update_order, columns=["time", "BLK_1", "BLK_2", "BLK_3", "BLK_4"])
df
```

```python
df.to_csv("./extended_order.csv", index=False)
```
